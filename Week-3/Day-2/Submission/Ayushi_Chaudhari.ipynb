```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("/content/ML-Bootcamp/Week-3/Day-2/Dominant color.jpg")
img =  cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
img = cv2.resize(img, (280,380))
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
x = img.reshape((-1,3))
x.shape
```

```python
from sklearn.cluster import KMeans
```

```python
k = 4
model = KMeans(n_clusters = k)
model.fit(x)
```

```python
centroids = model.cluster_centers_
```

```python
centroids
```

```python
import numpy as np
```

```python
colors = np.array(centroids,dtype='uint8')
```

```python
colors
```

```python
i = 1
for color in colors:
   plt.subplot(1,k,i)
   plt.axis("off")
   i = i + 1
   mat = np.zeros((100,100,3),dtype='uint8')
   mat[:,:,:] = color
   plt.imshow(mat)


plt.imshow(mat)
plt.show()
```
